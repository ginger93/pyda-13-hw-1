#### Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты.  
Принять и обосновать решение о дальнейшей работе с ними.

```python
q1 = df.age.quantile(0.25)
q3 = df.age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.age.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

Скорее всего выбросы по возрасту обоснованы некорректным заполнением данных. Возможно под 9 подразумевается лошади моложе 6 месяцев (2 в описании), но так как с точностью нельзя быть уверенным в этом, то данные принято удалить из выборки.

#### Удалить строки с пропусками будет не верным решением, поскольку удаляется одна треть выборки.

```python
remove_outliers.dropna().info()
```

На основании данных по результату (столбец outcome) можно заполнить столбцы пульс, переферический пульс и общий белок средними значениями, поскольку величина данных показателей могла напрямую повлиять на то выживет лошадь или нет.  
На основании переферического пульса можно заполнить показатели перельстатики и вздутия живота.

```python
fill_total_protein_median = remove_outliers
fill_total_protein_median.total_protein.fillna(df.groupby('outcome').total_protein.transform('median'), inplace=True)

fill_pulse_median = fill_total_protein_median
fill_pulse_median.pulse.fillna(df.groupby('outcome').pulse.transform('median'), inplace=True)

fill_mucous_membranes_median = fill_pulse_median
fill_mucous_membranes_median.mucous_membranes.fillna(df.groupby('outcome').mucous_membranes.transform('median'), inplace=True)

fill_peristalsis_median = fill_mucous_membranes_median
fill_peristalsis_median.peristalsis.fillna(df.groupby('mucous_membranes').peristalsis.transform('median'), inplace=True)

fill_abdominal_distension_median = fill_peristalsis_median
fill_abdominal_distension_median.abdominal_distension.fillna(df.groupby('mucous_membranes').abdominal_distension.transform('median'), inplace=True)
```

Поскольку остальные столбцы имеют незначительное количество пропусков ими можно пренебречь

```python
final = fill_abdominal_distension_median.dropna(thresh=8)
final
```

```python
for col in final.columns:
    pct_missing = final[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```
