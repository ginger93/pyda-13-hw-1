#### Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
rating = pd.read_csv('C:\\Users\\Admin\\Desktop\\netology\\Python\\Библиотека Pandas - продвинутый Пандас часть 2\\ml-latest-small\\ratings.csv')
rating_over100 = rating.groupby(['userId']).count().query('rating > 100').reset_index()
rating_over100 = rating_over100[['userId', 'rating']]

ltv_max = rating.groupby(['userId']).max().reset_index()
ltv_max = ltv_max[['userId', 'timestamp']]

ltv_min = rating.groupby(['userId']).min().reset_index()
ltv_min = ltv_min[['userId', 'timestamp']]

ltv = pd.merge(rating_over100, ltv_max, on = 'userId', how = 'left')

ltv_final = pd.merge(ltv, ltv_min, on = 'userId', how = 'left')

ltv_final['ltv'] = pd.to_datetime(ltv_final['timestamp_x'], unit='s') - pd.to_datetime(ltv_final['timestamp_y'], unit='s')
ltv_final.sort_values('ltv')
```
