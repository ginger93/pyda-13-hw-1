#### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('C:\\Users\\Admin\\Desktop\\netology\\Python\\Библиотека Pandas\\power.csv')
def baltic(country):
    if country in ['Latvia', 'Lithuania', 'Estonia']:
        return 'Прибалтика'
    return 'Другие'

power['baltic'] = power.country.apply(baltic)
power = power[power['baltic'] == 'Прибалтика']
power = power[(power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21)]
power = power[power['quantity'] >= 0]
power = power[(power['year'] == 2005) | (power['year'] == 2006) | (power['year'] == 2007)| (power['year'] == 2008) | (power['year'] == 2009) | (power['year'] == 2010)]
power = power['quantity'].sum()
print(power)
```
